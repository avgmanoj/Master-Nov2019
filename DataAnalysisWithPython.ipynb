```python
print(df.bachelor_s_degree_or_higher.mean())
print(df.geography.count())
print(df.geography_type.unique())
df.less_than_high_school_graduate.value_counts()
```

```python
# Method 1 
def mapGeography(x):
    if x == "City":
        return 1 
    else:
        return 0
# adding a new derived column to the data frame using a function
df['geography_mapped_value'] = df.geography_type.apply(mapGeography)

df.geography_mapped_value.value_counts() 
# new column value count
```

```python
# Method 2
#we can accomplish the same thing in a lambda function 
df['geography_mapped_value_lambda'] = df.geography_type.apply(lambda y: 1 if y == "City" else 0)

df.geography_mapped_value_lambda.value_counts()
```
